```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("advertising.csv")
```

```python
df
```

#To check if there is null value present or not

```python
df.isnull().sum()
```

# To view some basic statistical details like percentile, mean, std etc.

```python
df.describe()
```

# **Analysing relationship with Sales individually**

#Relationship Between TV and Sales

```python
plt.title("TV vs Sales")
plt.xlabel('TV')
plt.ylabel('Sales')

plt.scatter("TV", "Sales", data = df, c='g')
plt.show()
```

#Relationship Between Radio and Sales

```python
plt.title("Radio vs Sales")
plt.xlabel('Radio')
plt.ylabel('Sales')

plt.scatter("Radio", "Sales", data = df, c='g')
plt.show()
```

#Relationship Between Newspaper and Sales

```python
plt.title("Newspaper vs Sales")
plt.xlabel('Newspaper')
plt.ylabel('Sales')

plt.scatter("Newspaper", "Sales", data = df, c='g')
plt.show()
```

#Relationship with Sales together so that it would be easy to Analyse and Compare the plots

```python
sns.pairplot(df, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', size = 4)
```

# To show the percentage of a particular data from the whole

```python
cat1=["TV","Radio","Newspaper"]
tv=df['TV'].sum()
rd=df['Radio'].sum()
np=df['Newspaper'].sum()
cat2=[tv,rd,np]
colors=["y","r","g"]
plt.pie(cat2,labels=cat1,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=90)
plt.show()
```

# To estimate the probability density function of each one(TV , Radio & Newspaper)

```python
df.groupby(['Sales'])['TV','Radio','Newspaper'].mean().plot(kind='kde')
```

#To see both distribution of single variables and relationships between two variables

```python
sns.pairplot(df,diag_kind="kde")
```

#To represent the 'correlations' between pairs of variables in the given data.

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot = True)
plt.title("Correlation Matrix")
plt.show()
```

#To visualize the variation in the data distribution

```python
sns.distplot(df['Sales'])
```

# To visualize and analyze the relationship between two variables (Here, TV and Sales) and describe their individual distributions on the same plot.

```python
sns.jointplot(x='TV', y='Sales', data=df, kind='reg')
```
